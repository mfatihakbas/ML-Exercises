```python
penguins_df['sex'] = penguins_df['sex'].str.strip().str.lower().map({'male': 0, 'female': 1})
display(penguins_df.head())
```

Applied one-hot encoding to the 'sex' column and displayed the updated dataset.

```python
penguins_df['sex'] = penguins_df['sex'].fillna(penguins_df['sex'].mode()[0])
```

Filled NaN values in the 'sex' column with the mode value.

```python
penguins_df['bill_length_mm'] = penguins_df['bill_length_mm'].fillna(penguins_df['bill_length_mm'].mean())
penguins_df['bill_depth_mm'] = penguins_df['bill_depth_mm'].fillna(penguins_df['bill_depth_mm'].mean())
penguins_df['flipper_length_mm'] = penguins_df['flipper_length_mm'].fillna(penguins_df['flipper_length_mm'].mean())
penguins_df['body_mass_g'] = penguins_df['body_mass_g'].fillna(penguins_df['body_mass_g'].mean())
```

Filled NaN values in numerical columns with the mean value.

```python
display(penguins_df.isnull().sum())
display(penguins_df.head())
```

Checked for null values again to confirm they were filled properly.

```python
from sklearn.model_selection import train_test_split

label = penguins_df['species']
features = penguins_df.drop(columns=['species'])

X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=42)

print("\nEğitim ve Test Seti Boyutları:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```
